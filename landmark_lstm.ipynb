```python
import pandas as pd
df1=pd.read_excel('landmarks_data_combined_fall.xlsx')
df1['fall_label']=1
df1.head()
```

```python
df2=pd.read_excel('landmarks_data_combined_nofall.xlsx')
df2['fall_label']=0
df2.head()
```

```python
df=pd.concat([df1,df2],ignore_index=True)
df.head()
```

```python
data=df
```

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load your DataFrame (assumed to be pre-loaded as `df`)
# df contains rows for frames and columns like 'landmark_1_x', 'landmark_1_y', ..., 'landmark_33_x', 'landmark_33_y'

# Add video ID column to group frames by video (if not already present)
# df['video_id'] = (df['Frame'] == 1).cumsum()


# Feature columns
landmark_columns = [f'landmark_{i}_{axis}' for i in range(1, 34) for axis in ['x', 'y']]
target_column = 'fall_label'  # Target column indicating fall or no fall (binary)

# Prepare sequences for LSTM
X, y = [], []
for video_id, group in df.groupby('Folder'):
    if len(group) == 16:  # Ensure exactly 16 frames per video
        # Extract features and reshape
        X.append(group[landmark_columns].values)
        # Extract target (same label for all frames in a video)
        y.append(group[target_column].iloc[0])

X, y = np.array(X), np.array(y)
print(X.shape,y.shape)

# Split into train-test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Define LSTM Model
model = Sequential()
model.add(LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))  # Sigmoid for binary classification

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=16, validation_data=(X_test, y_test))

# Evaluate the model
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Save the model
model.save('fall_detection_lstm.h5')
```

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.regularizers import l2
from keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Prepare data (as in the original code)
landmark_columns = [f'landmark_{i}_{axis}' for i in range(1, 34) for axis in ['x', 'y']]
target_column = 'fall_label'

X, y = [], []
for video_id, group in df.groupby('Folder'):
    if len(group) == 16:
        X.append(group[landmark_columns].values)
        y.append(group[target_column].iloc[0])

X, y = np.array(X), np.array(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define LSTM Model with Regularization
model = Sequential()
model.add(LSTM(48, input_shape=(X_train.shape[1], X_train.shape[2]),
               kernel_regularizer=l2(0.01), return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid', kernel_regularizer=l2(0.01)))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Add Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history=model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test,y_test), callbacks=[early_stopping])

# Evaluate the model
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
