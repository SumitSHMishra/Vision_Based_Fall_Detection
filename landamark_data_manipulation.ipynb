```python
def extract_x_y_z_intensity(value):
    x=float(value.strip('()').split(',')[0])
    y=float(value.strip('()').split(',')[1])
    z=float(value.strip('()').split(',')[2])
    return x,y,z
    
```

```python
x,y,z=extract_x_y_z_intensity(k)
print(x,y,z)  
```

```python
import pandas as pd

# Function to extract x, y, z from a coordinate string

# Assuming 'df' is your DataFrame with 'landmark0' as a string of (x, y, z, intensity)

# Extract the z-coordinate for the head landmark
df['head_z'] = df['landmark0'].apply(lambda coords: extract_x_y_z_intensity(coords)[2])

# Calculate the difference in z-coordinates between consecutive frames
df['head_z_diff'] = df['head_z'].diff()

# Define the number of consecutive frames for which the head's z-distance must decrease
window_size = 2 # Set this based on how many frames you consider as "constant decrease"

# Create 'head_moved_closer' as True only if all differences in the window are negative
df['head_moved_closer'] = df['head_z_diff'].rolling(window=window_size).apply(
    lambda x: (x < 0).all(), raw=True
).fillna(False).astype(bool)

# Display the result
print(df[[ 'head_z', 'head_z_diff', 'head_moved_closer']])
```
